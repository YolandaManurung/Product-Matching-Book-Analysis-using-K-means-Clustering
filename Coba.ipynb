# Features Extraction

```python
import nltk
def tokenize_text(text):
    tokens = []
    for sent in nltk.sent_tokenize(text):
        for word in nltk.word_tokenize(sent):
            if len(word) < 2:
                continue
            tokens.append(word.lower())
    return tokens
genre = df.apply(
    lambda r: TaggedDocument(words=tokenize_text(r['genre']), tags=[r.book_title]), axis=1)
title = df.apply(
    lambda r: TaggedDocument(words=tokenize_text(r['book_title']), tags=[r.book_title]), axis=1)
```

```python
print("Genre:\n", genre.values[30])
print("\nTitle:\n", title.values[30])
```

```python
model_genre = Doc2Vec(dm=0, vector_size=300, negative=5, hs=0, min_count=2, sample = 0, workers=cores)
model_genre.build_vocab([x for x in tqdm(genre.values)])
```

```python
model_title = Doc2Vec(dm=0, vector_size=300, negative=5, hs=0, min_count=2, sample = 0, workers=cores)
model_title.build_vocab([x for x in tqdm(title.values)])
```
